```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='sin(x)')
plt.title('Пример линейного графика')
plt.xlabel('Значения x')
plt.ylabel('Значения y')
plt.legend()
plt.grid(True)
plt.show()
```

```python

np.random.seed(13)
N = 100
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (40 * np.random.rand(N))**2

plt.figure(figsize=(8, 6))
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.title('Пример диаграммы рассеяния')
plt.xlabel('Значения по x')
plt.ylabel('Значения по y')
plt.show()
```

```python
np.random.seed(2)
data = np.random.randn(1000)
plt.figure(figsize=(8, 6))
plt.hist(data, bins=100, alpha=0.9, color='red')
plt.title('Пример гистограммы нормально распределенных случайных данных')
plt.xlabel('Значения')
plt.ylabel('Частота')
plt.show()
```

```python

labels = ['pizza a', 'pie b', 'triangle C', 'Denis']
sizes = [15, 30, 45, 10]

plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightskyblue', 'lightgreen', 'gold'])
plt.axis('equal')
plt.title('Пример круговой диаграммы')
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)


plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)  # 2 строки, 1 столбец, первый график
plt.plot(x, y1, label='sin(x)', color='b')
plt.title('Графики синуса и косинуса')
plt.ylabel('Значения sin(x)')
plt.legend()

plt.subplot(2, 1, 2)  # 2 строки, 1 столбец, второй график
plt.plot(x, y2, label='cos(x)', color='r')
plt.xlabel('Значения x')
plt.ylabel('Значения cos(x)')
plt.legend()

plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax1 = plt.subplots(figsize=(10, 6))

ax2 = ax1.twinx()  # Создание второй оси y

ax1.plot(x, y1, 'g-', label='sin(x)')
ax1.set_xlabel('Значения x')
ax1.set_ylabel('sin(x)', color='g')
ax1.tick_params(axis='y', labelcolor='g')

ax2.plot(x, y2, 'b-', label='cos(x)')
ax2.set_ylabel('cos(x)', color='b')
ax2.tick_params(axis='y', labelcolor='b')

fig.suptitle('Группированные графики синуса и косинуса')
fig.legend(loc="upper right")
plt.show()
```

```python
import pandas as pd

# Создание данных временных рядов
date_rng = pd.date_range(start='2023-01-01', end='2023-01-10', freq='D')
data = np.random.randn(len(date_rng))

# Преобразование временных меток в формат Matplotlib
date = date_rng.to_pydatetime()

# Построение графика временного ряда
plt.figure(figsize=(10, 6))
plt.plot(date, data, marker='o')
plt.title('Пример визуализации временных рядов')
plt.xlabel('Дата')
plt.ylabel('Значение')
plt.show()
```
