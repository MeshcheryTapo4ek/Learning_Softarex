```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='sin(x)')
plt.title('  ')
plt.xlabel(' x')
plt.ylabel(' y')
plt.legend()
plt.grid(True)
plt.show()
```

```python

np.random.seed(13)
N = 100
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (40 * np.random.rand(N))**2

plt.figure(figsize=(8, 6))
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.title('  ')
plt.xlabel('  x')
plt.ylabel('  y')
plt.show()
```

```python
np.random.seed(2)
data = np.random.randn(1000)
plt.figure(figsize=(8, 6))
plt.hist(data, bins=100, alpha=0.9, color='red')
plt.title('     ')
plt.xlabel('')
plt.ylabel('')
plt.show()
```

```python

labels = ['pizza a', 'pie b', 'triangle C', 'Denis']
sizes = [15, 30, 45, 10]

plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightskyblue', 'lightgreen', 'gold'])
plt.axis('equal')
plt.title('  ')
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)


plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)  # 2 , 1 ,  
plt.plot(x, y1, label='sin(x)', color='b')
plt.title('   ')
plt.ylabel(' sin(x)')
plt.legend()

plt.subplot(2, 1, 2)  # 2 , 1 ,  
plt.plot(x, y2, label='cos(x)', color='r')
plt.xlabel(' x')
plt.ylabel(' cos(x)')
plt.legend()

plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax1 = plt.subplots(figsize=(10, 6))

ax2 = ax1.twinx()  #    y

ax1.plot(x, y1, 'g-', label='sin(x)')
ax1.set_xlabel(' x')
ax1.set_ylabel('sin(x)', color='g')
ax1.tick_params(axis='y', labelcolor='g')

ax2.plot(x, y2, 'b-', label='cos(x)')
ax2.set_ylabel('cos(x)', color='b')
ax2.tick_params(axis='y', labelcolor='b')

fig.suptitle('    ')
fig.legend(loc="upper right")
plt.show()
```
